```python
weights = []
for i in range(1,5):
    clf = MLPClassifier(hidden_layer_sizes=(2,2),
                        activation='logistic',
                        solver='adam',
                        learning_rate_init = 0.1,
                        random_state=i,
                        max_iter=1,
                        warm_start=True)
    for _ in range(2000):
        clf.fit(X, y)
        if i == 1: weights.append(clf.coefs_)
    draw_response(clf,X,y)
```

```python
import collections
def flatten(l):
    for el in l:
        if isinstance(el, collections.Iterable) and not isinstance(el, (str, bytes)):
            yield from flatten(el)
        else:
            yield el
            
weights_hist = [list(flatten(i)) for i in weights]
```
