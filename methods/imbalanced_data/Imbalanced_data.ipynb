Wyznaczaj wartości trzech metryk: tradycyjną trafność, G-mean oraz ROC AUC. W rozdziale 3.3.1.4 przeczytasz, jak uzyskać wartości dwóch (lub więcej) metryk podczas uruchomienia GridSearchCV. Użyj 10-fold stratified CV. 

```python
def gcv(scores, parameters, pipeline, X_train, X_test, y_train, y_test, nsplits=10, fn='ekhem', pred=False):
    for key, scorer in scores.items():
        print("# Tuning hyper-parameters for %s" % key)
        print()
        clf = GridSearchCV(pipeline,
                           param_grid=parameters,
                           scoring=scorer,
                           cv=StratifiedKFold(n_splits=nsplits),
                           n_jobs=-1)

        clf.fit(X_train, y_train)
        
        print("Best parameters set found on development set:")
        print()
        print(clf.best_params_)
        print()
        print("Grid scores on development set:")
        print()
        means = clf.cv_results_['mean_test_score']
        stds = clf.cv_results_['std_test_score']
        for mean, std, params in zip(means, stds, clf.cv_results_['params']):
            print("%0.3f (+/-%0.03f) for %r"
                  % (mean, std, params))
        print()
        if pred:
            print("Detailed classification report:")
            print()
            print("The model is trained on the full development set.")
            print("The scores are computed on the full evaluation set.")
            print()
            y_true, y_pred = y_test, clf.predict(X_test)
            print(classification_report(y_true, y_pred))
            print()
        dump(clf, "gs_results/gscv_{}_{}.joblib".format(str(key), fn))
```

```python
def get_acc_s(clf,title,filesave):
    annots={}
    df1_acc = pd.DataFrame(columns=['estimator', 'acc'])
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for i, (mean, std, params) in enumerate(zip(means, stds, clf.cv_results_['params'])):
        est = str(params['estimator']).split('(', 1)[0]
        df1_acc.loc[i] = [est,mean]
        annots[est] = "%0.3f (+/-%0.03f)" % (mean, std)
    heatmap1_data = pd.pivot_table(df1_acc, values='acc', index=['estimator'])
    annots2=[]
    for key, val in dict(heatmap1_data.to_dict()['acc']).items():
        annots2.append(annots[key])
    annots = np.reshape(annots2, (8, 1))
    ax = sns.heatmap(heatmap1_data, cmap="YlGnBu", annot=annots, fmt='s')
    ax.set_title(title)
#     plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")
#     plt.gcf().subplots_adjust(bottom=0.25)
    plt.tight_layout()
    plt.savefig(filesave)
    plt.plot()
    
def cm(clf, X_test, y_test, title, filesave):
    disp = plot_confusion_matrix(clf.best_estimator_, X_test, y_test,
                                 values_format = "d",
                                 cmap=plt.cm.Blues)
    print(clf.best_estimator_)
    print("ROC AUC: ",roc_auc_score(y_test, clf.best_estimator_.predict(X_test)))
    disp.ax_.set_title(title)
    plt.grid(False)
    plt.tight_layout()
    plt.savefig(filesave)
    plt.show()
```

```python
features = list(df.columns)
features.remove('class')
X = df.loc[:, features].values
y = df.loc[:,['class']].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

print(np.count_nonzero(y_test==0))
print(np.count_nonzero(y_train==0))

pipeline = Pipeline([('estimator', None)])

scores = {'accuracy': make_scorer(accuracy_score),
          'geometric_mean_score': make_scorer(geometric_mean_score),
          'roc': make_scorer(roc_auc_score)}

parameters = {'estimator':(KNeighborsClassifier(),
                           DecisionTreeClassifier(),
                           RandomForestClassifier(),
                           BalancedRandomForestClassifier(),
                           SVC(),
                           MLPClassifier(),
                           GaussianNB(),
                           QuadraticDiscriminantAnalysis())}

gcv(scores, parameters, pipeline, X_train, X_test, y_train, y_test, fn="7", pred=True)
```

Dla każdej z tych pięciu prób wyświetlaj macierz pomyłek (ze zbioru testowego uzyskanego dowolnym sposobem podziału). Czy widzisz wpływ class_weight w macierzach pomyłek? 

```python
features = list(df.columns)
features.remove('class')
X = df.loc[:, features].values
y = df.loc[:,['class']].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

scalar = PowerTransformer()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)

from sklearn.utils import class_weight
class_weights = class_weight.compute_class_weight('balanced',
                                                  np.unique(np.ravel(y_train,order='C')),
                                                  np.ravel(y_train,order='C'))
print(dict(enumerate(class_weights)))

class_weight_test = [
    {
        1: 0.5,
        0: 0.5
    },
    'balanced',
    {
        1: 0.3,
        0: 0.7
    },
    {
        1: 0.2,
        0: 0.8
    },
    {
        1: 0,
        0: 1
    },
]

for i, cwt_i in enumerate(class_weight_test):
    clf = DecisionTreeClassifier(class_weight = cwt_i, random_state=42)
    clf.fit(X_train, y_train)
    
    disp = plot_confusion_matrix(clf, X_test, y_test,
                                 values_format = "d",
                                 cmap=plt.cm.Blues)
    disp.ax_.set_title("Confusion matrix\n class_weight: {}".format(str(cwt_i)))
    plt.grid(False)
    plt.tight_layout()
    plt.savefig("cm_test_10_{}.pdf".format(i))
    plt.show()
```
